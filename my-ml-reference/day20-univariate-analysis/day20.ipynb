```python
# embarked ---> konse station se nikle
sns.countplot(df['Embarked'])
#df['Survived'].value_counts().plot(kind='bar')
```

# 2. Numerical Data

## a. Histogram

```python
import matplotlib.pyplot as plt
# use bins to control the ranges
plt.hist(df['Age'],bins=5)
```

## b. Distplot

KDE is an approximation on Probability density function (PDF)

```python
# sns.displot(df['Age'], kind='kde')
# sns.displot(df['Age'], kind='hist')

plt.figure(figsize=(8,5))

# alpha controls the opacity of the graph
sns.histplot(df['Age'], kde=True, stat="density", alpha=0.4)

plt.show()

# KDE --> kernel density estimation
```
